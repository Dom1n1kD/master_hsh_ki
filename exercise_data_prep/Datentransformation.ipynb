# Datentransformation

Im ersten Schritt werden die benötigten Pakete importiert

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('pauschalreisen.csv', delimiter = ';')
df
```

Nun sollen die kategorischen Werte in binäre umgewandelt werden. Dazu soll einmal die Pandas Funktion get_dummies und einmal der OneHotEncoder aus sklearn.preprocessing verwendet werden.

```python
df1=pd.get_dummies(df, columns=['Hotelgruppe','Fluggesellschaft '])
df1
```

```python
rf_enc = OneHotEncoder(handle_unknown='ignore')
X_encoded = rf_enc.fit_transform(df)#.toarray()
print(X_encoded)
```

Was ist der Unterschied im Output dieser beiden Möglichkeiten?

```python
df2=df1.copy()
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df2)
scaled_features = scaler.transform(df2)
df_scaled = pd.DataFrame(scaled_features, columns=df2.columns)
df_scaled.head()
```

Nun wollen wir die Spalte Preis p.P. in Intervalle einteilen und sie so in kategorische Werte umwandeln. Dafür sollen die pandas Funktionen qcut und cut verwendet werden.

## KL Hinweis:
Unterschied zwischen Cut und QCut (ChatGPT fragen)

```python
bin_labels_3 = ['niedrig', 'mittel', 'hoch']
df['Preis_bin'] = pd.qcut(df['Preis p.P.'],
                              q=[0, .33, .67, 1],
                              labels=bin_labels_3)
df
```

```python
bin_labels_3 = ['niedrig', 'mittel', 'hoch']
df['Preis_bin2'] = pd.cut(df['Preis p.P.'],
                              bins=3)#,labels=bin_labels_3)
df
```
